```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits , on='title')
```

```python
movies.shape
```

```python
# selecting the column which we required
movies = movies[['movie_id' , 'title' , 'overview' , 'genres' , 'keywords' , 'cast' , 'crew']]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# for converting string into list
import ast

def convert(object):
    L = []
    for i in ast.literal_eval(object):
        L.append(i['name'])
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(object):
    L = [];
    counter = 0;
    for i in ast.literal_eval(object) :
        if counter !=3:
            L.append(i['name']);
            counter = counter + 1;
        else:
         break; 


    return L
```

```python
# To get only top 3 actors from a film
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(object):
    L = [];
    for i in ast.literal_eval(object):
        if i['job'] == 'Director':
          L.append(i['name']);
          break;

    return L;
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
#removing spaces between words so that full name can be considered as single tag
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "") for i in x ])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "") for i in x ])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "") for i in x ])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "") for i in x ])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_df
```

```python
#Making single sentence
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.tags[0]
```

```python
#Selecting top 5000 most used words 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
#counting frequency of each word in the tags
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#for treating live,living,lives as single word
def stem(text):
    y = []
    
    for i in text.split(): 
       y.append(ps.stem(i)) 


    return " ".join(y) 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#calculates the cosine angle between each vectors from a particular vector
similarity = cosine_similarity(vectors)
```

```python
similarity[1633]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)) , reverse = True , key = lambda x:x[1])[1:6]




    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
